```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load tokenized data
train_path = "/content/tokenized_train_data.csv"
test_path = "/content/tokenized_test_data.csv"
val_path = "/content/tokenized_val_data.csv"

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
val_df = pd.read_csv(val_path)

# Convert tokenized strings into lists
train_df['Tokenized_Review'] = train_df['Tokenized_Review'].apply(eval)
test_df['Tokenized_Review'] = test_df['Tokenized_Review'].apply(eval)
val_df['Tokenized_Review'] = val_df['Tokenized_Review'].apply(eval)

# Tokenizer: Convert words to unique indices
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_df['Tokenized_Review'])

# Convert text to sequences (integer encoding)
train_sequences = tokenizer.texts_to_sequences(train_df['Tokenized_Review'])
test_sequences = tokenizer.texts_to_sequences(test_df['Tokenized_Review'])
val_sequences = tokenizer.texts_to_sequences(val_df['Tokenized_Review'])

# Padding to ensure uniform length
max_length = max(map(len, train_sequences))  # Find longest sequence
train_padded = pad_sequences(train_sequences, maxlen=max_length, padding='post')
test_padded = pad_sequences(test_sequences, maxlen=max_length, padding='post')
val_padded = pad_sequences(val_sequences, maxlen=max_length, padding='post')

# Load GloVe embeddings
def load_glove_embeddings(filepath, word_index, embedding_dim=100):
    vocab_size = len(word_index) + 1
    embedding_matrix = np.zeros((vocab_size, embedding_dim))

    with open(filepath, encoding="utf8") as f:
        for line in f:
            word, *vector = line.split()
            if word in word_index:
                idx = word_index[word]
                embedding_matrix[idx] = np.array(vector, dtype=np.float32)[:embedding_dim]

    return embedding_matrix

# Provide GloVe file path
glove_path = "/content/glove.6B.100d.txt"
embedding_dim = 100
embedding_matrix = load_glove_embeddings(glove_path, tokenizer.word_index, embedding_dim)

# Convert padded sequences into GloVe word embeddings
def get_embedding_representation(padded_sequences, embedding_matrix):
    embedded_reviews = []
    for seq in padded_sequences:
        embedded_seq = [embedding_matrix[word_idx] for word_idx in seq if word_idx != 0]  # Ignore padding (idx 0)
        embedded_reviews.append(np.mean(embedded_seq, axis=0) if embedded_seq else np.zeros(embedding_dim))
    return np.array(embedded_reviews)

train_embeddings = get_embedding_representation(train_padded, embedding_matrix)
test_embeddings = get_embedding_representation(test_padded, embedding_matrix)
val_embeddings = get_embedding_representation(val_padded, embedding_matrix)

# Convert embeddings to DataFrame and merge with original dataset
embedding_columns = [f"dim_{i}" for i in range(embedding_dim)]

train_emb_df = pd.DataFrame(train_embeddings, columns=embedding_columns)
test_emb_df = pd.DataFrame(test_embeddings, columns=embedding_columns)
val_emb_df = pd.DataFrame(val_embeddings, columns=embedding_columns)

train_final_df = pd.concat([train_df, train_emb_df], axis=1)
test_final_df = pd.concat([test_df, test_emb_df], axis=1)
val_final_df = pd.concat([val_df, val_emb_df], axis=1)

# Save the final embedded datasets
train_final_df.to_csv("train_embedded_data.csv", index=False)
test_final_df.to_csv("test_embedded_data.csv", index=False)
val_final_df.to_csv("val_embedded_data.csv", index=False)

print("GloVe embedding applied and saved successfully!")
```

```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# File Paths
train_path = "/content/tokenized_train_data.csv"
test_path = "/content/tokenized_test_data.csv"
val_path = "/content/tokenized_val_data.csv"
glove_path = "/content/glove.6B.100d.txt"  # GloVe embeddings

# Load Datasets
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
val_df = pd.read_csv(val_path)

# Convert Tokenized Strings into Lists
for df in [train_df, test_df, val_df]:
    df['Tokenized_Review'] = df['Tokenized_Review'].apply(eval)

# Tokenizer: Convert Words to Unique Indices
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_df['Tokenized_Review'])

# Convert Text to Sequences (Integer Encoding)
train_sequences = tokenizer.texts_to_sequences(train_df['Tokenized_Review'])
test_sequences = tokenizer.texts_to_sequences(test_df['Tokenized_Review'])
val_sequences = tokenizer.texts_to_sequences(val_df['Tokenized_Review'])

# Padding to Ensure Uniform Length
max_length = max(map(len, train_sequences))  # Find longest sequence
train_padded = pad_sequences(train_sequences, maxlen=max_length, padding='post')
test_padded = pad_sequences(test_sequences, maxlen=max_length, padding='post')
val_padded = pad_sequences(val_sequences, maxlen=max_length, padding='post')

# Load GloVe Embeddings
def load_glove_embeddings(filepath, word_index, embedding_dim=100):
    vocab_size = len(word_index) + 1
    embedding_matrix = np.zeros((vocab_size, embedding_dim))

    with open(filepath, encoding="utf8") as f:
        for line in f:
            word, *vector = line.split()
            if word in word_index:
                idx = word_index[word]
                embedding_matrix[idx] = np.array(vector, dtype=np.float32)[:embedding_dim]

    return embedding_matrix

embedding_dim = 100
embedding_matrix = load_glove_embeddings(glove_path, tokenizer.word_index, embedding_dim)

# Convert Padded Sequences into GloVe Word Embeddings
def get_embedding_representation(padded_sequences, embedding_matrix):
    embedded_reviews = []
    for seq in padded_sequences:
        embedded_seq = [embedding_matrix[word_idx] for word_idx in seq if word_idx != 0]  # Ignore padding (idx 0)
        embedded_reviews.append(np.mean(embedded_seq, axis=0) if embedded_seq else np.zeros(embedding_dim))
    return np.array(embedded_reviews)

train_embeddings = get_embedding_representation(train_padded, embedding_matrix)
test_embeddings = get_embedding_representation(test_padded, embedding_matrix)
val_embeddings = get_embedding_representation(val_padded, embedding_matrix)

# Convert Embeddings to DataFrame and Merge
embedding_columns = [f"dim_{i}" for i in range(embedding_dim)]
train_emb_df = pd.DataFrame(train_embeddings, columns=embedding_columns)
test_emb_df = pd.DataFrame(test_embeddings, columns=embedding_columns)
val_emb_df = pd.DataFrame(val_embeddings, columns=embedding_columns)

train_final_df = pd.concat([train_df, train_emb_df], axis=1)
test_final_df = pd.concat([test_df, test_emb_df], axis=1)
val_final_df = pd.concat([val_df, val_emb_df], axis=1)

# Add Sentiment Score Calculation
def compute_sentiment_score(df):
    df['text_sentiment'] = df['Processed_Review'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

    # Normalize Rating (Assuming rating is between 1 and 5)
    scaler = MinMaxScaler(feature_range=(1, 5))
    df['rating_scaled'] = scaler.fit_transform(df[['Rating']])

    # Convert Label to Numeric Sentiment (1 for real, 0 for fake)
    df['label_sentiment'] = df['Label'].apply(lambda x: 1 if x == "1" else 0)

    # Calculate Final Sentiment Score
    df['final_sentiment_score'] = (df['text_sentiment'] + df['rating_scaled'] + df['label_sentiment']) / 3

    return df

# Apply Sentiment Score Calculation
train_final_df = compute_sentiment_score(train_final_df)
test_final_df = compute_sentiment_score(test_final_df)
val_final_df = compute_sentiment_score(val_final_df)

# Save Final Data
train_final_df.to_csv("train_sentiment_data.csv", index=False)
test_final_df.to_csv("test_sentiment_data.csv", index=False)
val_final_df.to_csv("val_sentiment_data.csv", index=False)

print("GloVe embeddings and sentiment scores have been successfully added and saved!")
```

```python
import pandas as pd
import numpy as np
import spacy
import ast  # To convert stringified lists back to Python lists

# Load spaCy English Model
nlp = spacy.load("en_core_web_sm")

# Load Filtered Datasets
train_filtered = pd.read_csv("/content/train_filtered.csv")
test_filtered = pd.read_csv("/content/test_filtered.csv")
val_filtered = pd.read_csv("/content/val_filtered.csv")

# Load Tokenized Reviews
train_df = pd.read_csv("/content/tokenized_train_data.csv")
test_df = pd.read_csv("/content/tokenized_test_data.csv")
val_df = pd.read_csv("/content/tokenized_val_data.csv")
```

```python
# Function to extract POS ratios from tokenized words
def extract_pos_features(tokenized_words):
    # Convert stringified list back to Python list (if necessary)
    if isinstance(tokenized_words, str):
        tokenized_words = ast.literal_eval(tokenized_words)

    # Convert tokenized words into a spaCy Doc object
    doc = nlp(" ".join(tokenized_words))  # Reconstruct text for POS tagging
    total_words = len(doc)

    if total_words == 0:
        return pd.Series([0, 0, 0, 0])  # Avoid division by zero

    noun_count = sum(1 for token in doc if token.pos_ == "NOUN")
    verb_count = sum(1 for token in doc if token.pos_ == "VERB")
    adj_count = sum(1 for token in doc if token.pos_ == "ADJ")
    adv_count = sum(1 for token in doc if token.pos_ == "ADV")

    return pd.Series([
        noun_count / total_words,
        verb_count / total_words,
        adj_count / total_words,
        adv_count / total_words
    ])

# Apply POS feature extraction to Tokenized Reviews
train_pos = train_df["Tokenized_Review"].apply(extract_pos_features)
test_pos = test_df["Tokenized_Review"].apply(extract_pos_features)
val_pos = val_df["Tokenized_Review"].apply(extract_pos_features)

# Rename Columns
train_pos.columns = ["noun_ratio", "verb_ratio", "adj_ratio", "adv_ratio"]
test_pos.columns = ["noun_ratio", "verb_ratio", "adj_ratio", "adv_ratio"]
val_pos.columns = ["noun_ratio", "verb_ratio", "adj_ratio", "adv_ratio"]
```

```python
# Merge POS Features with Filtered Data
train_final = pd.concat([train_filtered, train_pos], axis=1)
test_final = pd.concat([test_filtered, test_pos], axis=1)
val_final = pd.concat([val_filtered, val_pos], axis=1)

# Save Updated Data
train_final.to_csv("train_final.csv", index=False)
test_final.to_csv("test_final.csv", index=False)
val_final.to_csv("val_final.csv", index=False)

print("Final datasets with POS features saved successfully!")
```
