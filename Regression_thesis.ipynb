```python
data=data.sample(frac=1)
data.head()
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Review_Content'] = data['Review_Content'].str.lower()
data['Review_Content'] = data['Review_Content'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
data['Review_Content'] = data['Review_Content'].apply(lambda x: re.sub(r'\s+', ' ', x))
data['Review_Content'] = data['Review_Content'].apply(lambda x: x.strip())
stop_words = set(stopwords.words('english'))
data['Review_Content'] = data['Review_Content'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
data.head(5)
```

```python
x=data['Review_Content']
y=data['Label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt=DT.predict(xv_test)
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB = GradientBoostingClassifier(random_state=0)
GB.fit(xv_train, y_train)
```

```python
pred_gb=GB.predict(xv_test)
GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(random_state=0)
RF.fit(xv_train, y_train)
```

```python
pred_rf=RF.predict(xv_test)
RF.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake Reviews"
    elif n == 1:
        return "Real Reviews"
def manual_testing(news):
    testing_news = {"Review_Content":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["Review_Content"] = new_def_test["Review_Content"].str.lower()
    new_def_test["Review_Content"] = new_def_test["Review_Content"].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
    new_def_test["Review_Content"] = new_def_test["Review_Content"].apply(lambda x: re.sub(r'\s+', ' ', x))
    new_def_test["Review_Content"] = new_def_test["Review_Content"].apply(lambda x: x.strip())
    stop_words = set(stopwords.words('english'))
    new_def_test["Review_Content"] = new_def_test["Review_Content"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
    new_x_test = new_def_test["Review_Content"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)
    return output_lable(pred_LR[0]),output_lable(pred_DT[0]),output_lable(pred_GB[0]),output_lable(pred_RF[0])
```
